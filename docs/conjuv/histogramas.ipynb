### df

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df_propostas = pd.read_csv("relatorio_31-10/data/propostas_conjuv.csv", sep=';')
df_propostas = df_propostas[(df_propostas['participatory_space/url'] == 'http://brasilparticipativo.presidencia.gov.br/assemblies/confjuv4?assembly_slug=confjuv4') & (~df_propostas['state'].isin(['rejected', 'withdrawn']))]

df_participantes = pd.read_csv("relatorio_31-10/data/participantes_conjuv.csv", sep=',', encoding='utf-8')

df_participantes2= pd.read_csv("relatorio_31-10/data/participantes_conjuv.csv", sep=',', encoding='ISO-8859-1')
df_geral = pd.read_csv("relatorio_31-10/data/delegados.csv", sep=';', encoding='utf-8')
df_delegados = pd.read_csv("relatorio_31-10/data/delegados2.csv", sep=';', encoding='utf-8')

```

### Votos

```python
hist, bin_edges = np.histogram(df_propostas['supports'], bins=15)

mediana = np.median(df_propostas['supports'])

plt.hist(df_propostas['supports'], bins=15, edgecolor='black', log=True)
plt.title('Distribuição de Votos por Propostas (Escala Logarítmica)')
plt.xlabel('Qtd de Votos')
#plt.ylabel('Propostas')
plt.axvline(mediana, color='red', linestyle='dashed', linewidth=2, label=f'Mediana: {mediana}')
plt.yticks([])
plt.legend()

plt.show()
```

### comentários 

```python

hist, bin_edges = np.histogram(df_propostas['comments'], bins=15)
mediana_comments = np.median(df_propostas['comments'])

plt.hist(df_propostas['comments'], bins=15, edgecolor='black', log=True)
plt.title('Distribuição de Comentários por Propostas')
plt.xlabel('Qtd de Comentários')
#plt.ylabel('Propostas')
plt.yticks([])
plt.axvline(mediana_comments, color='red', linestyle='dashed', linewidth=2, label=f'Mediana: {mediana_comments}')
plt.legend()

plt.show()
```
